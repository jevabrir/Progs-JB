```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.2f}'.format
```

####**2. Obtencin de Datos de Produccin**

```python
# @title
# --------------------------------------------
# 1 GENERACIN DE DATOS (puedes ocultar esta celda)
# --------------------------------------------

np.random.seed(42)

dias = 20
plan = 1000

# Generar variables base
tiempo_paro = np.random.randint(5, 70, dias)
operadores = np.random.choice([3,4,5], dias, p=[0.2,0.3,0.5])

# Modelo realista de produccin
# Produccin depende negativamente del tiempo de paro
# y positivamente del nmero de operadores

produccion_base = (
    plan
    - (tiempo_paro * 5)          # cada min de paro reduce 5 unidades
    + (operadores * 15)          # ms operadores ayudan
)

# Agregamos variabilidad controlada
produccion_real = produccion_base + np.random.normal(0, 20, dias)

# Limitar valores lgicos
produccion_real = np.clip(produccion_real, 750, 1000)

# Defectos proporcionales a produccin (2% a 6%)
defectos = produccion_real * np.random.uniform(0.02, 0.06, dias)

data = pd.DataFrame({
    "Dia": range(1, dias+1),
    "Produccion_Plan": plan,
    "Tiempo_Paro": tiempo_paro,
    "Operadores": operadores,
    "Produccion_Real": produccion_real.astype(int),
    "Defectos": defectos.astype(int)
})
```

####**6. Anlisis de Tendencia y Prediccin**  

Ejecuta las siguientes lneas de cdigo para la creacin del:  

*  **Modelo lienal para estimacin de la Eficiencia(%)**

El cdigo incluye comentarios explicativos para facilitar su comprensin.

```python
# Bloque 1: Clculo del modelo lineal Tiempo de Paro vs. Eficiencia

# Regresin lineal: m = pendiente , b = intercepto
m, b = np.polyfit(
    data["Tiempo_Paro"],        # Variable independiente (X)
    data["Eficiencia (%)"],     # Variable dependiente (Y)
    1)                          # Grado 1: modelo lineal

# Calcular correlacin y coeficiente de determinacin R
correlacion = data["Tiempo_Paro"].corr(data["Eficiencia (%)"])
r2 = correlacion**2

# Imprimir la ecuacin del modelo y R
print("Modelo lineal estimado:")
print(f"Eficiencia = {m:.2f} * Tiempo_Paro + {b:.2f}")
print(f"R = {r2:.3f}")
```

Ejecuta las siguientes lneas de cdigo para la creacin de la **Grfica de "Tiempo de Paro vs Eficiencia (%) con el modelo".**

El cdigo incluye comentarios explicativos para facilitar su comprensin.

```python
plt.figure(figsize=(11, 3.5))    # Crea la figura con tamao (ancho, alto en pulgadas)

# Grfica de dispersin: Eje X, Eje Y, etiqueta
plt.scatter(data["Tiempo_Paro"],data["Eficiencia (%)"],alpha=0.7,
    label="Datos reales")

# Genera valores ordenados de X para una lnea suave
x_vals = np.linspace(data["Tiempo_Paro"].min(),data["Tiempo_Paro"].max(),100)

# Calcula los valores estimados usando el modelo
y_vals = m * x_vals + b

# Dibuja la lnea de tendencia
plt.plot(x_vals,y_vals,linewidth=3,label="Lnea de tendencia")

# Identifica das crticos con eficiencia menor al 95%
criticos = data[data["Eficiencia (%)"] < 95]

# Resalta los das crticos: Eje X, Eje Y, etiqueta
plt.scatter(criticos["Tiempo_Paro"],criticos["Eficiencia (%)"],s=100,
    label="Das Crticos (<95%)")

# Escribe la ecuacin y R dentro de la grfica
ecuacion = f"Eficiencia = {m:.2f}x + {b:.2f}\nR = {r2:.2f}"
plt.text(0.95, 0.95,ecuacion,
    transform=plt.gca().transAxes,
    verticalalignment='top',
    horizontalalignment='right',
    bbox=dict(boxstyle="round", alpha=0.3))

plt.title("Impacto del Tiempo de Paro en la Eficiencia")  # Ttulo
plt.xlabel("Tiempo de Paro (min)")                         # Eje X
plt.ylabel("Eficiencia (%)")                               # Eje Y
plt.legend()                                               # Leyenda
plt.grid(alpha=0.3)                                        # Cuadrcula
```

Ejecuta las siguientes lneas de cdigo para la creacin de una **Funcin para estimar la Eficienca del Proceso en funcin de Tiempo de Paro**. El cdigo incluye comentarios explicativos para facilitar su comprensin.

```python
def estimar_eficiencia(tiempo_paro):
    # Esta funcin recibe como entrada el tiempo de paro (en minutos)
    # y utiliza el modelo lineal (pendiente m y ordenada b)
    # para estimar la eficiencia del proceso.

    eficiencia = m * tiempo_paro + b  # Frmula de la recta: y = mx + b

    # Mostrar el resultado en pantalla
    print(f"Si el tiempo de paro es {tiempo_paro} minutos,")
   # print(f"La eficiencia estimada es {eficiencia:.2f}%")

    # Regresa el valor calculado por si se quiere usar despus
    return eficiencia
```

Ejecuta las siguientes lneas de cdigo para la **Estimar la Eficienca del Proceso en funcin de Tiempo de Paro**. El cdigo incluye comentarios explicativos para facilitar su comprensin.

```python
# Guardamos el resultado en una variable para usarlo despus
ef_Estimada = estimar_eficiencia(30)

# Mostramos el valor almacenado
print(f"Eficiencia estimada es: {ef_Estimada:.2f}%")
```
